```python
car = {"brand": "Toyota", "model": "Corolla", "year": 2020}
print(car["year"])
```

```python
person = {"name": "Alice", "age": 25}
person["city"] = "Bogota"
print(person)
```

Añadimos un dato asignando un valor a una clave nueva

```python
person = {"name": "Marilyn", "age": 25, "city": "Bogota"}
person["age"] = 16
print(person)
```

Se accede a la clave "age" y se le da un nuevo valor

```python
person = {"name": "Alice", "age": 25, "city": "Bogota"}
del person["city"]
print(person)
```

Usamos la palabra .pop() para quitar "city".

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
print(fruits.keys())
```

.keys() nos devuelve la lista de las frutas.

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
print(fruits.values())
```

.values() para ver las cantidades

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
for clave, valor in fruits.items():
    print(f"{clave}: {valor}")
```

Usamos "for" e ".items()" para obtener clave y valor

```python
grades = {"Ana": 85, "Luis": 90, "Maria": 78}
print("Luis" in grades)
```

Se usa in para comprobar si la clave existe

```python
grades = {"Ana": 85, "Luis": 90, "Maria": 78}
print("Pedro" in grades)
```

```python
countries = {"Colombia": "Bogota", "France": "Paris", "Japan": "Tokyo"}
for pais, capital in countries.items():
    print(f"La capital de {pais} es {capital}")
```

Recorre el diccionario mostrando información de forma ordenada.

```python
temperatures = {"Monday": 22, "Tuesday": 24, "Wednesday": 21, "Thursday": 23}
print(max(temperatures.values()))
```
